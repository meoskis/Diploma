```python
import pandas as pd
import numpy as np
```

```python
# То, что собиралось вручную + отдельным кравлером
data_1 = pd.read_excel('/content/drive/MyDrive/ДИПЛОМ/Списки компаний/ogtn_added_data.xlsx')
data_1 = data_1.rename(columns={'Название':'excel_name', 'ИНН':'inn'})
data_1.head()
```

```python
# Все собиралось кравлером
data_2 = pd.read_excel('/content/drive/MyDrive/ДИПЛОМ/Списки компаний/ogrn_added_data_2.xlsx')
data_2 = data_2.rename(columns={'Название':'excel_name'})
data_2['interfax_name'] = np.NaN
data_2.head()
```
