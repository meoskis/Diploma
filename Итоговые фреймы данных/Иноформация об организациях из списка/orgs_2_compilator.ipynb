```python
import json

# with open(path+'/data_main_okved_1.json') as f:
#     main_1 = json.load(f)

# with open(path+'/data_main_okved_2.json') as f:
#     main_2 = json.load(f)

with open(path+'/data_main_okved_3.json') as f:
    main_3 = json.load(f)




# with open(path+'/data_second_okved_1.json') as f:
#     second_1 = json.load(f)

# with open(path+'/data_second_okved_2.json') as f:
#     second_2 = json.load(f)

with open(path+'/data_second_okved_3.json') as f:
    second_3 = json.load(f)
```

```python
main_okveds =  main_3
second_okveds = second_3
```

```python
main_okveds_id = []
main_okveds_name = []
for org in main_okveds:
  try:
    key = list(org.keys())[0]
    value = org[key]
    main_okveds_id.append(key)
    main_okveds_name.append(value)
  except IndexError:
    main_okveds_id.append(np.nan)
    main_okveds_name.append(np.nan)
```

```python
second_okveds_all = []
for org in second_okveds:
  try:
    key = list(org.keys())[0]
    value = org[key]
    second_okveds_all.append({key:value})
  except IndexError:
    second_okveds_all.append(np.nan)
```

```python
oveds = pd.DataFrame()
oveds['main_okved_id'] = main_okveds_id
oveds['main_okved_name'] = main_okveds_name
oveds['second_okveds'] = second_okveds_all
```
