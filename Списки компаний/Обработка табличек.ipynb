```python
data = pd.read_excel('/content/drive/MyDrive/ДИПЛОМ/Списки компаний/Рейтинг крупнейших компаний России по объему реализации продукции RAEX-600.xlsx',
                     index_col=0)
display(data.head())
display(data.shape)
display(data.columns)
```

```python
data.isna().sum()
```

```python
chart_data = pd.DataFrame(Counter(data['Отрасль']).most_common())
chart_data.columns = ['отрасль', 'количество']
chart_data.index = chart_data['отрасль']
chart_data.drop(['отрасль'], axis=1, inplace=True)
chart_data
```

```python
colors = sns.color_palette('pastel')[0:5]

#create pie chart
plt.pie(chart_data['количество'], labels = chart_data['количество'].index, colors = colors, autopct='%.0f%%')
plt.show()
```
